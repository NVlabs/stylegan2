```python
import os
import numpy as np
import PIL.Image
import dnnlib
import dnnlib.tflib as tflib
import pretrained_networks
import os
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
%matplotlib inline
```

```python
OUT_DIR = 'out/'
OUT_IMG_DIR = OUT_DIR+'img/'
```

```python
tflib.init_tf()
```

```python
network = '/data/yoder_lab/networks/stylegan2_256x256_balanced.pkl'
```

```python
_G, _D, Gs = pretrained_networks.load_networks(network)
```

```python
Gs_syn_kwargs = dnnlib.EasyDict()
Gs_syn_kwargs.output_transform = dict(func=tflib.convert_images_to_uint8, nchw_to_nhwc=True)
Gs_syn_kwargs.randomize_noise = False
Gs_syn_kwargs.minibatch_size = 1
```

```python
def make_latent_z(Gs, seed=0, sigma = 1.0, mu = 0.0):
    latent =  sigma * np.random.RandomState(seed).randn(1, Gs.input_shape[1]) + mu
    return latent
```

```python
latents = np.zeros((100_000,512))
```

```python
for seed in range(0,latents.shape[0]):
    latents[seed,:] = make_latent_z(Gs,seed)
```

```python
if not os.path.isdir(OUT_DIR):
    os.mkdir(OUT_DIR)
if not os.path.isdir(OUT_IMG_DIR):
    os.mkdir(OUT_IMG_DIR)
```

```python
np.save(OUT_DIR+'/latents.npy',latents)
```

```python
latents[0,:].shape
```

```python
for seed in range(0,latents.shape[0]):
    Gs_output = Gs.run(latents[seed:seed+1,:], None, **Gs_syn_kwargs) # [seed, y, x, rgb]
    image = PIL.Image.fromarray(Gs_output[0], 'RGB')
    filename = OUT_IMG_DIR+'/'+str(seed)+'.png'
    print('Saving',filename)
    image.save(filename)
```
